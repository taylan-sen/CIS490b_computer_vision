# Introduction to k-means

This notebook goes through a 1d example followed by a 2d example of k-means applied to the "2 sizes fit all" scenario from the clustering lesson.

```python
# import necessary modules
import sklearn
import numpy as np
import pandas as pd

print('imports successful')
```

```python
# set up the shoe size data as the merger of two bell curves
dist1 = np.random.normal(11.5, 1.5, 1000)
df1 = pd.DataFrame(dist1)
df1.columns = ['size']
dist2 = np.random.normal(6.5, 1.5, 1000)
df2 = pd.DataFrame(dist2)
df2.columns = ['size']

df_all = pd.concat([df1, df2])

# visualize the distributions
ax= df1.hist(bins=20, alpha=0.5)
df2.hist(bins=20, ax=ax, alpha=0.5)
df_all.hist(bins=20)
```

```python
import matplotlib.pyplot as plt
# use k-means from the sklearn module to find the two shoe sizes
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=2).fit(df_all)

kmeans.cluster_centers_
plt
plt.hist(df_all)
plt.grid(True)
plt.ylabel('frequency')
plt.xlabel('input')
plt.scatter(kmeans.cluster_centers_[:,0], [0,0], color='red', s=100)
plt.show()
```

**2D Example**

```python
# establish two distributions, both as dist1size = np.random.normal(11.5, 1.5, 1000)
dist1width = np.random.normal(6, 0.5, 1000)
df1 = pd.DataFrame(zip(dist1size, dist1width))
df1.columns=['size','width']

dist2size = np.random.normal(7, 1.5, 1000)
dist2width = np.random.normal(4, 0.5, 1000)
df2 = pd.DataFrame(zip(dist2size, dist2width))
df2.columns=['size','width']

df_all = pd.concat([df1, df2])

#display(df1)
ax = df1.plot.scatter('size', 'width', color='r')
df2.plot.scatter('size', 'width', ax=ax)
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=2).fit(df_all)

kmeans.cluster_centers_
```

```python
df_means = pd.DataFrame(kmeans.cluster_centers_)
df_means.columns = ['size','width']
display(df_means)

ax = df1.plot.scatter('size', 'width', color='r')
df2.plot.scatter('size', 'width', ax=ax)
df_means.plot.scatter('size','width',ax=ax,color='y', s=100)
```
