```python
import cv2
import matplotlib.pyplot as plt

imgL = cv2.imread('left.png',0)
imgR = cv2.imread('right.png',0)
print(imgL.shape)
print(imgR.shape)
stereo_obj = cv2.StereoBM.create(numDisparities=16,blockSize=15)
disp_map = stereo_obj.compute(imgL, imgR)

plt.imshow(disp_map)
plt.show()
```
