## dlib
###fast facial landmark detectorlibrary

This library will allow us to extract facial landmarks...

![](https://b2633864.smushcdn.com/2633864/wp-content/uploads/2017/04/facial_landmarks_68markup-1024x825.jpg)

Facial landmarks are a useful building block for head pose extraction, emotion estimation, as well as many other human image related tasks.

```python
import dlib
import cv2
import matplotlib.pyplot as plt

img = cv2.imread('image.jpg')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY) # convert to grayscale
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
detector = dlib.get_frontal_face_detector()
rects = detector(gray, 1) # rects contains all the faces detected
print(rects)
for rect in rects:
  cv2.rectangle(img,
                (rect.tl_corner().x,rect.tl_corner().y),
                (rect.br_corner().x,rect.br_corner().y),
                (255, 0, 255), 5)
plt.imshow(img)
```

```python
img = dlib.load_rgb_image('image.jpg')
#img = cv2.imread('image.jpg')
#gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY) # convert to grayscale
#img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
detector = dlib.get_frontal_face_detector()
predictor = dlib.shape_predictor('shape_predictor_68_face_landmarks.dat')
rects = detector(img, 1) # rects contains all the faces detected
print(rects)
for rect in rects:
  cv2.rectangle(img,
                (rect.tl_corner().x,rect.tl_corner().y),
                (rect.br_corner().x,rect.br_corner().y),
                (255, 0, 255), 5)
  shape = predictor(img,rect)
  print(shape.num_parts)
  print("Part 0: {}, Part 1: {} ...".format(shape.part(0),
                                            shape.part(1)))
  for p in shape.parts():
    cv2.circle(img, (p.x, p.y), 20, (255,255,255), -1)

plt.imshow(img)
```
